```python
#Load Librari
import numpy as np
import matplotlib.pyplot as plt
import cv2
from skimage import data
```

```python
image = data.page()
plt.subplot(1, 3,1)
plt.hist(image.ravel(),255,[0,255])
mask = image >= 70 #seting jika kurang dari 70 jadi 255
image[mask]=255

# hitam = image > 187
# image[hitam]=0


plt.figure(num=None, figsize=(15, 20), dpi=100)

plt.subplot(1,3,2),plt.imshow(data.page(),cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(1,3,3),plt.imshow(image,cmap = 'gray')
plt.title('masked'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
from PIL import ImageEnhance
from PIL import Image, ImageFilter
# Read image
im = Image.open('Foto/rijal.jpeg')
enh = ImageEnhance.Contrast(im)
hasil = enh.enhance(10.0)

plt.figure(num=None, figsize=(20, 30), dpi=100)
plt.subplot(1, 2, 1), plt.imshow(im)
plt.title('Original')
plt.subplot(1, 2, 2), plt.imshow(hasil)
plt.title('erosion')
plt.show()
```
