```python
#load Librarinya
import numpy as np
import cv2
from matplotlib import pyplot as plt
import os
```

```python
folder_path = "D:\Kuliah\Semester 4\Computer Vesion\Tugas\Foto_noise"

# Fungsi Untuk Menampilkan Gambar
def TampilGambar(img,filname):  
    plt.imshow(img),plt.title(filname)
     
# Membuat fungsi untuk menampilkan isi folder nya
def listDir(dir):
    # Ambil isi file yang ada di folder_path
    fileName = os.listdir(dir)
    # Looping fileName nya
    for fileNames in fileName :
        path = os.path.abspath(os.path.join(dir, fileNames))
        img = cv2.imread(path)
        Averaging = cv2.blur(img,(10,10))
        Gaussian = cv2.GaussianBlur(img,(5,5),10)
        Median = cv2.medianBlur(img,(3))
        Bilateral = cv2.bilateralFilter(img,5,300,150)

        plt.subplot(121),TampilGambar(img,'Gambar Original')
        plt.xticks([]),plt.yticks([])
        plt.subplot(122),TampilGambar(Averaging,'Filter dengan Metode Averaging')
        plt.xticks([]),plt.yticks([])
        plt.show()

        plt.subplot(121),TampilGambar(img,'Gambar Original')
        plt.xticks([]),plt.yticks([])
        plt.subplot(122),TampilGambar(Gaussian,'Filter dengan Metode Gaussian')
        plt.xticks([]),plt.yticks([])
        plt.show()

        plt.subplot(121),TampilGambar(img,'Gambar Original')
        plt.xticks([]),plt.yticks([])
        plt.subplot(122),TampilGambar(Median,'Filter dengan Metode Median')
        plt.xticks([]),plt.yticks([])
        plt.show()

        plt.subplot(121),TampilGambar(img,'Gambar Original')
        plt.xticks([]),plt.yticks([])
        plt.subplot(122),TampilGambar(Bilateral,'Filter dengan Metode Bilateral')
        plt.xticks([]),plt.yticks([])
        plt.show()

if __name__ == "__main__" :
    listDir(folder_path)
```
