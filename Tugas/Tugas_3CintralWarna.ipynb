```python
import numpy as np
import cv2
import os
from matplotlib import pyplot as plt
```

```python
# Masukan path folder yang ada gambar nya
folder_path = "D:\Kuliah\Semester 4\Computer Vesion\Latihan dan Materi\Foto Doraemon"

#Fungsi untuk Load Gambar
def LoadGambar(img):
    return cv2.imread(img)
# Fungsi untuk Tampil Plot
def TampilPlot(img,ruang_warna):
    return cv2.cvtColor(img, ruang_warna)
    
#Fungsi untuk menampilkan Gambar
def TampilGambar (img,fileName):
    plt.imshow(img)
    plt.title(fileName)
    plt.show()

# Fungsi untuk Mengolah warna gambar
def PengolahanCitra (img,filename,color):
    image = LoadGambar(img)
    r = image.copy()
    # set green and blue channels to 0
    r[:, :, 1] = 0
    r[:, :, 2] = 0

    g = image.copy()
    # set blue and red channels to 0
    g[:, :, 0] = 0
    g[:, :, 2] = 0

    b = image.copy()
    # set red and green channels to 0
    b[:, :, 1] = 0
    b[:, :, 2] = 0

    plt.figure(num=None, figsize=(15, 20), dpi=100)
    for i,col in enumerate(color) :
        histr = cv2.calcHist([image], [i], None, [256], [0, 256])
        plt.subplot(4, 2, 1)
        plt.plot(histr, color=col)
        plt.xlim([0, 256])
        plt.title('Histogram citra asli')
        Rgb = cv2.COLOR_BGR2RGB
        Lab = cv2.COLOR_BGR2LAB

    plt.subplot(4, 2, 2), TampilGambar(TampilPlot(r,Rgb),' Ruang Warna RGB')
    plt.subplot(4, 2, 2), TampilGambar(TampilPlot(b,Lab),' Ruang Warna LAB')
    plt.subplot(4, 3, 1), TampilGambar(TampilPlot(b,cv2.COLOR_RGB2HSV),' Ruang Warna      HSV')
    
# Membuat fungsi untuk menampilkan isi folder nya
def listDir(dir):
    # Ambil isi file yang ada di folder_path
    fileName = os.listdir(dir)
    # Looping fileName nya
    for fileNames in fileName :
        path = os.path.abspath(os.path.join(dir, fileNames))
        color = ('r')
        PengolahanCitra(path,fileNames,color)

if __name__ == "__main__" :
    listDir(folder_path)

```
