```python
import cv2
from matplotlib import pyplot as plt
import numpy as np


def loadgambar():
    return cv2.imread('Foto/1.jpeg')


def tampil_plot(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)


def figure():
    return plt.figure(num=None, figsize=(15, 20), dpi=100)
```

```python
image = loadgambar()
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
r = image.copy()
# set green and blue channels to 0
r[:, :, 1] = 37
r[:, :, 2] = 134

g = image.copy()
# set blue and red channels to 0
g[:, :, 0] = 0
g[:, :, 2] = 0

b = image.copy()
# set red and green channels to 0
b[:, :, 1] = 0
b[:, :, 2] = 0

color = ('r', 'g', 'b')
plt.figure(num=None, figsize=(15, 20), dpi=100)
for i, col in enumerate(color):
    histr = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.subplot(4, 2, 1)
    plt.plot(histr, color=col)
    plt.xlim([0, 256])
    plt.title('Histogram citra asli')
plt.subplot(4, 2, 2), plt.imshow(image), plt.title('citra asli')

color = ('r')
for i, col in enumerate(color):
    histr = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.subplot(4, 2, 3)
    plt.plot(histr, color=col)
    plt.xlim([0, 256])
    plt.title('Histogram citra red')
plt.subplot(4, 2, 4), plt.imshow(tampil_plot(r)), plt.title('red')

color = ('g')
for i, col in enumerate(color):
    histr = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.subplot(4, 2, 5)
    plt.plot(histr, color=col)
    plt.xlim([0, 256])
    plt.title('Histogram citra green')
plt.subplot(4, 2, 6), plt.imshow(tampil_plot(g)), plt.title('green')

color = ('b')
for i, col in enumerate(color):
    histr = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.subplot(4, 2, 7)
    plt.plot(histr, color=col)
    plt.xlim([0, 256])
    plt.title('Histogram citra blue')
plt.subplot(4, 2, 8), plt.imshow(tampil_plot(b)), plt.title('blue')
plt.show()  # To show figure
```

```python
image = loadgambar()
dim = (100, 100)

# resize image
image = cv2.resize(image, dim, interpolation=cv2.INTER_AREA)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
# lakukan thresholdreshold pada citra grey
ret, bw_img = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

figure()
plt.subplot(1, 3, 1), plt.imshow(
    tampil_plot(image)), plt.title('Original image')
plt.subplot(1, 3, 2), plt.imshow(tampil_plot(rgb)), plt.title('Gray image')
plt.subplot(1, 3, 3), plt.imshow(tampil_plot(bw_img)), plt.title('BW image')
plt.show()
print('original \n', image)
print('grayscale \n', gray)
print('bw \n', bw_img)
```

```python
img = cv2.imread('Foto/1.jpeg', 0)
ret, thresh = cv2.threshold(img, 125, 255, cv2.THRESH_BINARY)
ret, thresh_inv = cv2.threshold(img, 125, 255, cv2.THRESH_BINARY_INV)
ret, thresh_trunc = cv2.threshold(img, 125, 255, cv2.THRESH_TRUNC)
ret, thresh_tozero = cv2.threshold(img, 125, 255, cv2.THRESH_TOZERO)
ret, thresh_tozero_inv = cv2.threshold(img, 125, 255, cv2.THRESH_TOZERO_INV)


cv2.imshow('Thresh binary', thresh)
cv2.imshow('Thresh binary inv', thresh_inv)
cv2.imshow('Thresh trunc', thresh_trunc)
cv2.imshow('Thresh to zero', thresh_tozero)
cv2.imshow('Thresh to zero inv', thresh_tozero_inv)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
