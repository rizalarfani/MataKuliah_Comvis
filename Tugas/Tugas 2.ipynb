```python
# Masukan path folder yang ada gambar nya
folder_path = "D:\Kuliah\Semester 4\Computer Vesion\Dasar Python\Foto Doraemon"

# Membuat fungsi untuk menampilkan isi folder nya
def listDir(dir) :
    # Ambil isi file yang ada di folder_path
    fileName = os.listdir(dir)
    # Looping fileName nya
    for fileNames in fileName :
        path = os.path.abspath(os.path.join(dir, fileNames))
        img = cv2.imread(path)
        TampilGambar(img,fileNames)

def KecilGambar(gambar) :
    # Membuat variable
    h,e = gambar.shape[:2]
    # menentukan ukuran gambar yang mau di kecilin
    new_h, new_e = int(h/2),int(e/2)
    return cv2.resize(gambar, (new_h,new_e))

def TampilGambar(img,filename) :
    resize = KecilGambar(img)
    plt.imshow(resize)
    plt.title(filename)
    plt.show()

    # Menampilkan liwat Window
    cv2.imshow(filename,resize)
    cv2.waitKey()
    cv2.destroyAllWindows() 

    # Cek jika folder_path tidak ditemukan
if __name__ == "__main__" :
    listDir(folder_path)
```
